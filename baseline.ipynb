```python
from datasets import load_dataset
import os
from PIL import Image

# 데이터셋 다시 다운로드 (스트리밍 방식)
ds = load_dataset("ahmed-ai/skin-lesions-classification-dataset", streaming=True)

# 디렉토리 생성
base_dir = "./data/MIT_skin-lesions-classification-dataset"
os.makedirs(base_dir, exist_ok=True)

train_dir = os.path.join(base_dir, "train")
test_dir = os.path.join(base_dir, "test")

os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

def save_images(dataset, split, save_dir):
    for i, example in enumerate(dataset[split]):
        # PIL 이미지를 직접 저장
        image = example['image']
        label = example['label']
        label_dir = os.path.join(save_dir, str(label))
        os.makedirs(label_dir, exist_ok=True)
        image_path = os.path.join(label_dir, f"{split}_{i}.jpg")
        image.save(image_path)
        if i % 100 == 0:
            print(f"Saved {i} images from {split} set")

# 이미지 저장
save_images(ds, 'train', train_dir)
save_images(ds, 'test', test_dir)
```

```python
import os
import pandas as pd
import shutil

base_dir = "./data/MIT_skin-lesions-classification-dataset"

# 레이블과 질병명 매핑 딕셔너리
label_to_disease = {
    0: 'Actinic keratoses',
    1: 'Basal cell carcinoma',
    2: 'Benign keratosis-like-lesions',
    3: 'Chickenpox',
    4: 'Cowpox',
    5: 'Dermatofibroma',
    6: 'Healthy',
    7: 'HFMD',
    8: 'Measles',
    9: 'Melanocytic nevi',
    10: 'Melanoma',
    11: 'Monkeypox',
    12: 'Squamous cell carcinoma',
    13: 'Vascular lesions',
}

# 디렉토리 구조 파악 및 이미지 이동 함수
def move_images_and_create_csv(base_dir, split):
    image_paths = []
    labels = []
    
    for label_dir in os.listdir(os.path.join(base_dir, split)):
        label_index = int(label_dir)
        label_name = label_to_disease[label_index]
        label_path = os.path.join(base_dir, split, label_dir)
        
        for img_name in os.listdir(label_path):
            src_path = os.path.join(label_path, img_name)
            dest_path = os.path.join(base_dir, split, img_name)
            shutil.move(src_path, dest_path)
            image_paths.append(f'{split}/{img_name}')
            labels.append(label_name)
    
    df = pd.DataFrame({'image_path': image_paths, 'label': labels})
    df.to_csv(os.path.join(base_dir, f'{split}.csv'), index=False)

# train 및 test 디렉토리에서 이미지 이동 및 CSV 파일 생성
move_images_and_create_csv(base_dir, 'train')
move_images_and_create_csv(base_dir, 'test')
```
