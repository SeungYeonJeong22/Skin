```python
import pandas as pd
import matplotlib.pyplot as plt

#  
train_df = pd.read_csv('./data/modified_AICamp/train.csv')
valid_df = pd.read_csv('./data/modified_AICamp/valid.csv')
test_df = pd.read_csv('./data/modified_AICamp/test.csv')

#   
train_counts = train_df['label'].value_counts().sort_index()
valid_counts = valid_df['label'].value_counts().sort_index()
test_counts = test_df['label'].value_counts().sort_index()

# 
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

axs[0].bar(train_counts.index.astype(str), train_counts.values, color='skyblue')
axs[0].set_title("Train Set Label Distribution")
axs[0].set_ylabel("Count")
axs[0].tick_params(axis='x', rotation=45)

axs[1].bar(valid_counts.index.astype(str), valid_counts.values, color='lightgreen')
axs[1].set_title("Validation Set Label Distribution")
axs[1].set_ylabel("Count")
axs[1].tick_params(axis='x', rotation=45)

axs[2].bar(test_counts.index.astype(str), test_counts.values, color='salmon')
axs[2].set_title("Test Set Label Distribution")
axs[2].set_ylabel("Count")
axs[2].set_xlabel("Label")
axs[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
